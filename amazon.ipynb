```python
categorical_columns = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority']
for col in categorical_columns:
    print(f"{col} unique values:\n{df[col].unique()}\n")
```

```python
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(df2[col])
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime
df2['Order Date'] = pd.to_datetime(df2['Order Date'])
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'])

# Extract year and month from 'Order Date'
df2['Order Year'] = df2['Order Date'].dt.year
df2['Order Month'] = df2['Order Date'].dt.month

# Plot sales over time
plt.figure(figsize=(14, 7))
df2.groupby('Order Date')['Total Revenue'].sum().plot()
plt.title('Total Revenue Over Time')
plt.ylabel('Total Revenue')
plt.show()
```

```python
df2
```

```python
df2['Total Revenue']=df2['Total Revenue'].astype(int)
df2[ 'Unit Price']=df2[ 'Unit Price'].astype(int)
df2[ 'Unit Cost']=df2['Unit Cost'].astype(int)
df2['Total Cost']=df2['Total Cost'].astype(int)
df2[ 'Total Profit']=df2[ 'Total Profit'].astype(int)
```

```python
df2
```

```python
# Distribution of numerical columns
numerical_columns = ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df2[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
corr = df2[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#Total revenue by Regions
plt.figure(figsize=(15,10))
sns.barplot(data=df2,x='Region',y='Total Revenue')
plt.show()
```
