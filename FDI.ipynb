```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_melted=df.melt(id_vars=['Sector'],var_name='Year',value_name='FDI')
df_melted['Year']=df_melted['Year'].apply(lambda x:int(x.split('-')[0]))
df2=df_melted.head(10)
```

```python
sector_data=df_melted[df_melted['Sector']=='METALLURGICAL INDUSTRIES']
sns.lineplot(x='Year',y='FDI',data=sector_data,marker='o')
plt.title('FDI trend for Metallurgical industries (2000-2017)')
plt.xlabel('year')
plt.ylabel('FDI(Billion USD)')
plt.grid(True)
plt.show()
```

```python
sector_data1=df_melted[df_melted['Sector']=='MINING']
sns.lineplot(x='Year',y='FDI',data=sector_data1,marker='o')
plt.title('FDI trend for MINING (2000-2017)')
plt.xlabel('year')
plt.ylabel('FDI(Billion USD)')
plt.grid(True)
plt.show()
```

```python
sector_data2=df_melted[df_melted['Sector']=='POWER']
sns.lineplot(x='Year',y='FDI',data=sector_data2,marker='o')
plt.title('FDI trend for POWER (2000-2017)')
plt.xlabel('year')
plt.ylabel('FDI(Billion USD)')
plt.grid(True)
plt.show()
```

```python
##COMPARING SOME SECTORS##
```

```python
sector_to_compare=['METALLURGICAL INDUSTRIES','MINING','POWER']
filtered_data=df_melted[df_melted['Sector'].isin(sector_to_compare)]
filtered_data
```

```python
sns.lineplot(x='Year',y='FDI',hue='Sector',data=filtered_data,marker='o')
plt.title('FDI comaprison for selected sectors (2000-2017)')
plt.xlabel('Year')
plt.ylabel('FDI(Billion USD)')
plt.show()
```
