```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Attrition data.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df2=df.head(100)
df2
```

```python
df2.columns
```

```python
df.describe()
```

```python
sns.countplot(x='Attrition',data=df2)
plt.title('Attrition Distribution')
plt.show()
```

```python
categorical_features=['Gender','JobRole','Department','MaritalStatus']
for features in categorical_features:
    plt.figure(figsize=(15,10))
    sns.countplot(x=features,data=df2,hue='Attrition')
    plt.title(f'{features} vs Attrition')
    plt.show()
```

```python
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='Attrition',y='MonthlyIncome',data=df2)
plt.title('Attrition vs Monthly Income')
plt.show()
```

```python
sns.boxplot(x='JobRole',y='MonthlyIncome',hue='Attrition',data=df2)
plt.xticks(rotation=90)
plt.title('Job role vs Monthly Income vs Attrition')
plt.show()
```

```python
sns.boxplot(x='Attrition',y='Age',hue='Attrition',data=df2)
plt.xticks(rotation=90)
plt.title('Age vs Attrition')
plt.show()
```

```python
sns.violinplot(x='Attrition',y='TotalWorkingYears',data=df2)
plt.title('Total Working Years vs Attrition')
plt.show()
```

```python
corr_matrix=df2[['Age','MonthlyIncome','YearsAtCompany','TotalWorkingYears']].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('correlation matrix')
plt.show()
```

```python
sns.countplot(x='EnvironmentSatisfaction',hue='Attrition',data=df2)
plt.title('Environment Satisfaction vs Attrition')
plt.show()
```

```python
sns.countplot(x='JobSatisfaction',hue='Attrition',data=df2)
plt.title('Job Satisfaction vs Attrition')
plt.show()
```

```python
sns.countplot(x='WorkLifeBalance',hue='Attrition',data=df2)
plt.title('Wrok Life Balance vs Attrition')
plt.show()
```
