```python
df2=df.head(10)
df2
```

```python
print(df2['sex'].value_counts())
print(df2['cp'].value_counts())
print(df2['slope'].value_counts())
print(df2['ca'].value_counts())
print(df2['fbs'].value_counts())
print(df2['restecg'].value_counts())
print(df2['thal'].value_counts())
print(df2['target'].value_counts())
print(df2['exang'].value_counts())
```

```python
print(df2[['age','trestbps','chol','thalach','oldpeak']].describe())
```

```python
for col in ['sex','trestbps','chol','thalach','oldpeak']:
    sns.countplot(x=col,data=df2)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
numeric_cols=['age','trestbps','chol','thalach','oldpeak']
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df2[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
corr=df2.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
```

```python
sns.countplot(x='target',data=df2)
plt.show()
```

```python
for col in ['age','trestbps','chol','thalach','oldpeak']:
    sns.boxplot(x='target',y=col,data=df2)
    plt.show()
```

```python
sns.pairplot(df2[['age','trestbps','chol','thalach','oldpeak','target']],hue='target')
plt.show()
```
